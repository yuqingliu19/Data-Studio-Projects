```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_excel('1960.xlsx', skiprows=16)
df
```

```python
df.sort_values(by='Population').tail(10).plot(kind='barh', x='Region, subregion, country or area *', y='Population', legend=False)
plt.tight_layout()
plt.savefig("1960.pdf")
```
