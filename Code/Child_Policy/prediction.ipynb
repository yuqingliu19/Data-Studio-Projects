```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_male = pd.read_excel('POPULATION_BY_AGE_MALE.xlsx', sheetname = 'MEDIUM VARIANT', skiprows=16)
df_male
```

```python
df_CM = df_male[df_male['Region, subregion, country or area *']=='China']
```

```python
df_CM.drop(df_CM.columns[[0, 1, 2, 3, 4]], axis=1, inplace=True)
```

```python
df_CM.rename(columns={'Reference date (as of 1 July)': 'Year'}, inplace=True)
df_CM
```

```python
df_2020M = df_CM[df_CM['Year'] == 2020].transpose().reset_index()
df_2020M
```

```python
df_2020M.rename(columns={'index': 'Age', 1387: '2020'}, inplace=True)
df_2020M
```

```python
df_2020M.drop(df_2020M.index[0], inplace=True)
df_2020M
```
